```python
#!/usr/bin/python
# -*- coding: latin-1 -*-
import emcee
# from cdecimal import Decimal
import decimal
from astropy.io import ascii
import sys
# import novas
import math
import numpy as np
#import triangle
import matplotlib
import matplotlib.pyplot as plt
#get_ipython().magic('matplotlib')           # osx')
plt.rcParams['font.size']=10
from astropy.coordinates import SkyCoord
from astropy.time import Time
import astropy.units as u
from astropy import wcs
from astropy.io import fits
import astropy.coordinates as coord
# from novas import compat as novas
# from novas.compat import eph_manager
from matplotlib.legend_handler import HandlerLine2D
import time
from pylab import *

pi_g = np.arccos(-1)
nimg = 16
ndim = 5 + nimg*4 +1
nwalkers = 150
ncores = 64
delta_coord = 50   #(in mas)
```

```python
full_cat_x = np.load('full_cat_x_2019.npy')
full_cat = np.load('full_cat_2019.npy')
epoch = np.load('epoch_2019.npy')

nstars = len(full_cat_x)
ra_med = np.zeros(nstars)
dec_med = np.zeros(nstars)
fix = {}
for i in range(nstars):
    ra_all_l = []
    dec_all_l = []
    for q in range(len(full_cat[i])-1):
        if q%3 == 0:
            ra_all_l.append(full_cat[i,1+q])
            dec_all_l.append(full_cat[i,2+q])
        ra_all = np.array(ra_all_l)
        dec_all = np.array(dec_all_l)
        ra_med[i] = np.median(ra_all[np.where(ra_all != 0)])
        if len(ra_all[np.where(ra_all != 0)]) < 2:
            ra_med[i] = 0.
        dec_med[i] = np.median(dec_all[np.where(dec_all != 0)])
        if len(dec_all[np.where(dec_all != 0)]) < 2:
            dec_med[i] = 0.
        fix['ra_star'+str(i)] = ra_med[i]*u.deg.to(u.marcsec)
        fix['dec_star'+str(i)] = dec_med[i]*u.deg.to(u.marcsec)


#convert catalogue
data = np.zeros((nimg,nstars,3))
for i in range(nimg):
    for j in range(nstars):
        data[i,j,0] = full_cat_x[j,1+i*3]
        data[i,j,1] = full_cat_x[j,2+i*3]
        data[i,j,2] = full_cat_x[j,3+i*3]


guess = {'ra_g':np.float64(14.4952634805555556)*15.*u.deg.to(u.marcsec),
         'dec_g':np.float64(-62.67948975)*u.deg.to(u.marcsec),
         'mu_ra_g':-3775.75,
         'mu_dec_g':765.54,
         'pi_g':768.13,
         'jitter_g':1.0}

for i in range(nimg):
    guess['theta_g'+str(i)] = 0.
    guess['off_center_x_g'+str(i)] = 0.
    guess['off_center_y_g'+str(i)] = 0.
    guess['rho_g'+str(i)] = 12.25

def vector2params(vec):
    params = {}
    params['ra'] = vec[0]  #RA Proxima
    params['dec'] = vec[1]
    params['mu_ra'] = vec[2]
    params['mu_dec'] = vec[3]
    params['pi'] = vec[4]
    for i in range(nimg):
        params['theta'+str(i)] = vec[5+i*4]
        params['rho'+str(i)] = vec[6+i*4]
        params['off_center_x'+str(i)] = vec[7+i*4]
        params['off_center_y'+str(i)] = vec[8+i*4]
    params['jitter'] = vec[41]
    return params

def params2vector(params):
    vec = np.zeros(len(params.keys()))
    vec[0] = params['ra']  #RA Proxima
    vec[1] = params['dec']
    vec[2] = params['mu_ra']
    vec[3] = params['mu_dec']
    vec[4] = params['pi']
    for i in range(nimg):
        vec[5+i*4] = params['theta'+str(i)]
        vec[6+i*4] = params['rho'+str(i)]
        vec[7+i*4] = params['off_center_x'+str(i)]
        vec[8+i*4] = params['off_center_y'+str(i)]
    vec[41] = params['jitter']
    return vec

def guess2vector(guess):
    guess_vec = np.zeros(len(guess.keys()))
    guess_vec[0] = guess['ra_g']
    guess_vec[1] = guess['dec_g']
    guess_vec[2] = guess['mu_ra_g']
    guess_vec[3] = guess['mu_dec_g']
    guess_vec[4] = guess['pi_g']
    for i in range(nimg):
        guess_vec[5+i*4] = guess['theta_g'+str(i)]
        guess_vec[6+i*4] = guess['rho_g'+str(i)]
        guess_vec[7+i*4] = guess['off_center_x_g'+str(i)]
        guess_vec[8+i*4] = guess['off_center_y_g'+str(i)]
    guess_vec[41] = guess['jitter_g']
    return guess_vec

guess_vec = guess2vector(guess)
p0 = np.zeros((nwalkers,ndim))
for i in range(nwalkers):
    p0[i,:] = np.random.normal(guess_vec,1e-4,ndim)
```

```python
def lnprior(params):
    for key in params.keys():
        if key == 'mu_ra':
            if params[key] < (guess['mu_ra_g']-delta_coord) or params[key] > (guess['mu_ra_g']+delta_coord):
                #print('{} {}'.format(key, params[key]))
                return -np.inf
        elif key == 'mu_dec':
            if params[key] < (guess['mu_dec_g']-delta_coord) or params[key] > (guess['mu_dec_g']+delta_coord):
                #print('{} {}'.format(key, params[key]))
                return -np.inf
        elif key == 'pi':
            if params[key] < (guess['pi_g']-delta_coord) or params[key] > (guess['pi_g']+delta_coord):
                #print('{} {}'.format(key, params[key]))
                return -np.inf
        elif 'theta' in key:
            if params[key] < -1. or params[key] > 1.:
                #print('{} {}'.format(key, params[key]))
                return -np.inf
        elif 'off_center' in key:
            if params[key] < -0.5 or params[key] > 0.5:
                #print('{} {}'.format(key, params[key]))
                return -np.inf
        elif 'rho' in key:
            if params[key] < 11 or params[key] > 13:
                #print('{} {}'.format(key, params[key]))
                return -np.inf
        elif 'ra' in key:
            if params[key] < (guess['ra_g']-delta_coord) or params[key] > (guess['ra_g']+delta_coord):
                #print('{} {}'.format(key, params[key]))
                return -np.inf
        elif 'dec' in key:
            if params[key] < (guess['dec_g']-delta_coord) or params[key] > (guess['dec_g']+delta_coord):
                #print('{} {}'.format(key, params[key]))
                return -np.inf
        elif 'jitter' in key:
            if params[key] < 0. or params[key] > 3.:
                #print('{} {}'.format(key, params[key]))
                return -np.inf
        else:
            return 0.0
    return 0.0

def lnprob(vec,data):
    params = vector2params(vec)
    lp = lnprior(params)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(params,data)

jd_start, jd_end, number = eph_manager.ephem_open()
acc = 0
model_cat = novas.make_cat_entry ('Proxima','x',acc,guess['ra_g']/(15.*u.deg.to(u.marcsec)),guess['dec_g']/(u.deg.to(u.marcsec)),guess['mu_ra_g'],guess['mu_dec_g'],guess['pi_g'],-22.20)
prova = np.zeros((nimg,nstars,2))
for i in range(nimg):
    model_ra_h, model_dec_deg = novas.astro_star(epoch[i], model_cat,acc)
    cos_prox = np.cos(model_dec_deg*u.deg.to(u.rad))
    theta = guess['theta_g'+str(i)]*u.deg
    for j in range(nstars):

        #step 1: offset with respect to the center of the Proxima
        delta_ra = -fix['ra_star'+str(j)] + model_ra_h*15.*u.deg.to(u.marcsec)  #ra(mas)
        delta_ra_mas = delta_ra*cos_prox  #ra(mas)
        delta_dec_mas = fix['dec_star'+str(j)] - model_dec_deg*u.deg.to(u.mas) #dec(mas)

        #step 2: rot for the TN correction
        delta_x_mas = delta_ra_mas*np.cos(theta.to(u.rad)) - delta_dec_mas*np.sin(theta.to(u.rad))
        delta_y_mas = delta_ra_mas*np.sin(theta.to(u.rad)) + delta_dec_mas*np.cos(theta.to(u.rad))

        #step 3: pixel scale adjustment

        delta_x_px = delta_x_mas/guess['rho_g'+str(i)] + guess['off_center_x_g'+str(i)]
        delta_y_px = delta_y_mas/guess['rho_g'+str(i)] + guess['off_center_y_g'+str(i)]

        prova[i,j,0] = delta_x_px
        prova[i,j,1] = delta_y_px

'''
ima = 0
ind=abs(data[ima,:,0])>0
figure()
plot(data[ima,:,0][ind],data[ima,:,1][ind],'or')
plot(prova[ima,:,0][ind],prova[ima,:,1][ind],'og',mfc='none',ms=10)
'''

def lnlike(params,data):
    #print params['ra']
    #print params['mu_ra']
    jd_start, jd_end, number = eph_manager.ephem_open()
    acc = 0
    model_cat = novas.make_cat_entry ('Proxima','x',acc,params['ra']/(15.*u.deg.to(u.marcsec)),params['dec']/(u.deg.to(u.marcsec)),params['mu_ra'],params['mu_dec'],params['pi'],-22.20)
    pos = np.zeros((nimg,nstars,2))
    for i in range(nimg):
        model_ra_h, model_dec_deg = novas.astro_star(epoch[i], model_cat,acc)
        cos_prox = np.cos(model_dec_deg*u.deg.to(u.rad))
        theta = params['theta'+str(i)]*u.deg
        for j in range(nstars):

            #step 1: offset with respect to the center of the Proxima
            delta_ra = -fix['ra_star'+str(j)] + model_ra_h*15.*u.deg.to(u.mas)  #ra(mas)
            delta_ra_mas = delta_ra*cos_prox  #ra(mas)
            delta_dec_mas = fix['dec_star'+str(j)]- model_dec_deg*u.deg.to(u.mas) #dec(mas)

            #step 2: rot for the TN correction, rotation is counterclockwise!
            delta_x_mas = delta_ra_mas*np.cos(-theta.to(u.rad)) - delta_dec_mas*np.sin(-theta.to(u.rad))
            delta_y_mas = delta_ra_mas*np.sin(-theta.to(u.rad)) + delta_dec_mas*np.cos(-theta.to(u.rad))

            #step 3: pixel scale adjustment

            delta_x_px = delta_x_mas/params['rho'+str(i)] + params['off_center_x'+str(i)]
            delta_y_px = delta_y_mas/params['rho'+str(i)] + params['off_center_y'+str(i)]

            #print delta_x_px

            pos[i,j,0] = delta_x_px
            pos[i,j,1] = delta_y_px

    #likehood
    #data = data
    res_norm = []
    ey = []
    for i in range(nimg):
        for j in range(nstars):
            #for z in range(2):
            if data[i,j,0] != 0 and data[i,j,1] != 0:
                iey = np.sqrt((np.sqrt(2.)*data[i,j,2])**2 + params['jitter']**2)
                ey.append(iey)
                #print (data[i,j,z] - pos[i,j,z])
                res_norm.append((np.sqrt((data[i,j,0] - pos[i,j,0])**2. + (data[i,j,1] - pos[i,j,1])**2.))/iey)
    ey =np.array(ey)
    res_norm = np.array(res_norm)
    loglike = -0.5*(res_norm**2 + np.log(2.0*pi_g*ey**2))
    return np.sum(loglike)


t1 = time.time()
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=(data,), threads=ncores)
runs = 130000

sampler.run_mcmc(p0, runs)
print('computing time {:.3f} h'.format((time.time()-t1)/3600.))

np.save('result_jose_fix_2019.npy', sampler.chain)
np.save('prob_jose_fix_2019.npy', sampler.lnprobability)


'''
chainz = plt.figure(figsize = (15, 12))
for i in range(ndim):
    plt.plot(sampler.chain[:,:,i].T, '-', color='k', alpha=0.3)
chainz.savefig('chainz.png')
plt.show(block=False)
sys.exit()

N = 100
aux = sampler.chain[:,N:,:].reshape((-1,ndim))
post_distr_fun = corner.corner(aux, quantiles=(0.1587, 0.5, 0.8413), levels=(1-np.exp(-0.5), 1-np.exp(-2), 1-np.exp(-4.5)), show_titles=True,title_fmt=".5f", title_args={"fontsize": 12}, \
	plot_datapoints=False, fill_contours=True, smooth=1.0, bins=20.)
post_distr_fun.savefig('post_jose.pdf')
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt


def logprob(x):
    return np.sum(
        np.logaddexp(-0.5 * (x - 2) ** 2, -0.5 * (x + 2) ** 2,)
        - 0.5 * np.log(2 * np.pi)
        - np.log(2)
    )


x = np.linspace(-5.5, 5.5, 5000)
plt.plot(x, np.exp(list(map(logprob, x))), "k")
plt.yticks([])
plt.xlim(-5.5, 5.5)
plt.ylabel("p(x)")
plt.xlabel("x");
```
